### Import Required Libraries and Set Up Environment Variables

```python
# Dependencies
import requests
import time
from dotenv import load_dotenv
import os
import pandas as pd
import json
```

```python
# Set environment variables from the .env in the local environment
load_dotenv()

nyt_api_key = os.getenv("NYT_API_KEY")
tmdb_api_key = os.getenv("TMDB_API_KEY")
```

### Access the New York Times API

```python
# Set the base URL
url1 = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"

# Filter for movie reviews with "love" in the headline
# section_name should be "Movies"
# type_of_material should be "Review"
filter_query = 'section_name:"Movies" AND type_of_material:"Review" AND headline:"love"'

# Use a sort filter, sort by newest
sort = "newest"

# Select the following fields to return:
# headline, web_url, snippet, source, keywords, pub_date, byline, word_count
field_list = "headline,web_url,snippet,source,keywords,pub_date,byline,word_count"

# Search for reviews published between a begin and end date
begin_date = "20130101"
end_date = "20230531"

# Build URL
nyt_url = url1 + "fq=" + filter_query + "&sort=" + sort + "&fl=" + field_list + "&begin_date=" + begin_date + "&end_date=" + end_date + "&api-key=" + nyt_api_key
nyt_url
```

```python
# Create an empty list to store the reviews
empty_list = []

# loop through pages 0-19
for i in range(20):

    # create query with a page number
    # API results show 10 articles at a time
    page = str(i)
    nyt_url_page = nyt_url + "&page=" + page
    
    # Make a "GET" request and retrieve the JSON
    nyt_response = requests.get(nyt_url_page)
    nyt_resp_j = nyt_response.json()

    # empty_list.append(nyt_resp_j["response"]["docs"][0])
    # print(empty_list)

    # Add a twelve second interval between queries to stay within API query limits
    time.sleep(12)
    
    # Try and save the reviews to the reviews_list
    try:
        # loop through the reviews["response"]["docs"] and append each review to the list
        for resp in nyt_resp_j["response"]["docs"]:
            empty_list.append(resp)
            
        # Print the page that was just retrieved
        print(f"Loaded page " + page)

        # Print the page number that had no results then break from the loop
    except:
        print(f"No records on page" + page)
        break
```

```python
# Preview the first 5 results in JSON format
# Use json.dumps with argument indent=4 to format data
first_5 = empty_list[:5]
json_review_1 = json.dumps(first_5, indent=4)
print(json_review_1)
```

```python
# Convert reviews_list to a Pandas DataFrame using json_normalize()
empty_list_df = pd.json_normalize(empty_list)
empty_list_df
```

```python
# Extract the title from the "headline.main" column and
# save it to a new column "title"
# Title is between unicode characters \u2018 and \u2019. 
# End string should include " Review" to avoid cutting title early
empty_list_df["title"] = empty_list_df["headline.main"].apply(lambda substring: substring[substring.find("\u2018")+1: substring.find("\u2019 Review")])
empty_list_df
```

```python
# Extract 'name' and 'value' from items in "keywords" column
def extract_keywords(keyword_list):
    extracted_keywords = ""
    for item in keyword_list:
        # Extract 'name' and 'value'
        keyword = f"{item['name']}: {item['value']};" 
        # Append the keyword item to the extracted_keywords list
        extracted_keywords += keyword
    return extracted_keywords

# Fix the "keywords" column by converting cells from a list to a string
empty_list_df["keywords"] = empty_list_df["keywords"].apply(extract_keywords)
empty_list_df
```

```python
# Create a list from the "title" column using to_list()
# These titles will be used in the query for The Movie Database
titles_for_later = empty_list_df["title"].to_list()
titles_for_later
```

### Access The Movie Database API

```python
# Prepare The Movie Database query
url2 = "https://api.themoviedb.org/3/search/movie?query="
tmdb_key_string = "&api_key=" + tmdb_api_key
tmdb_key_string2 = "?api_key=" + tmdb_api_key
url3 = "https://api.themoviedb.org/3/movie/"
```

```python
# Create an empty list to store the results
another_empty_list = []

# Create a request counter to sleep the requests after a multiple
# of 50 requests
counter = 0

# Loop through the titles
for title in titles_for_later:
    # Check if we need to sleep before making a request
    if counter % 50 == 0:
        time.sleep(1)

    # Add 1 to the request counter
    counter += 1
    
    # Perform a "GET" request for The Movie Database
    
    # Include a try clause to search for the full movie details.
    # Use the except clause to print out a statement if a movie
    # is not found.
    try:
        # Get movie id
        tmdb_response = requests.get(url2 + title + tmdb_key_string)
        tmdb_resp_j = tmdb_response.json()
        # print(json.dumps(tmdb_resp_j, indent=4))
        movie_id = str(tmdb_resp_j["results"][0]["id"])
        # print(movie_id)

        # Make a request for a the full movie details
        

        # Execute "GET" request with url
        # print(f"{url3}{movie_id}?api_key={tmdb_api_key}")
        # print(f"{url3}{movie_id}{tmdb_key_string2}")
        # new_url = url3 + movie_id + tmdb_key_string2
        # print(new_url)

        tmdb_response2 = requests.get(url3 + movie_id + tmdb_key_string2)
        tmdb_resp_j_full = tmdb_response2.json()
        # print(json.dumps(tmdb_resp_j_full, indent=4))
        

        # Extract the genre names into a list
        genre_list = []
        for genre in tmdb_resp_j_full["genres"]:
            genre_list.append(genre["name"])
        
        # print(genre_list)

        # Extract the spoken_languages' English name into a list
        language_list = []
        for language in tmdb_resp_j_full["spoken_languages"]:
            language_list.append(language["english_name"])

        # print(language_list)

        # Extract the production_countries' name into a list
        prod_countries = []
        for countries in tmdb_resp_j_full["production_countries"]:
            prod_countries.append(countries["name"])

        # print(prod_countries)

        # Add the relevant data to a dictionary and
        # append it to the tmdb_movies_list list
        detail_dict = {
            "title": title,
            "original_title": tmdb_resp_j_full["original_title"],
            "budget": tmdb_resp_j_full["budget"],
            "original_language": tmdb_resp_j_full["original_language"],
            "homepage": tmdb_resp_j_full["homepage"],
            "overview": tmdb_resp_j_full["overview"],
            "popularity": tmdb_resp_j_full["popularity"],
            "runtime": tmdb_resp_j_full["runtime"],
            "revenue": tmdb_resp_j_full["revenue"],
            "release_date": tmdb_resp_j_full["release_date"],
            "vote_average": tmdb_resp_j_full["vote_average"],
            "vote_count": tmdb_resp_j_full["vote_count"],
            "genres": genre_list,
            "spoken_languages": language_list,
            "production_countries": prod_countries 
        }
        another_empty_list.append(detail_dict)

        # Print out the title that was found
        print(f"{title} was added to another hecken' good list")
    except:
        print(f"{title} was not found")

```

### Merge and Clean the Data for Export

```python
# Merge the New York Times reviews and TMDB DataFrames on title
merged_df = pd.merge(another_empty_list_df,empty_list_df, on="title")
pd.set_option('display.max_columns', None)
merged_df.head(5)
```

```python
# Remove list brackets and quotation marks on the columns containing lists
# Create a list of the columns that need fixing
janky_columns = ['genres', 'production_countries', 'spoken_languages']

# Create a list of characters to remove
shady_characters = ["[","]","'"]

# Loop through the list of columns to fix
for column in janky_columns:
    # Convert the column to type 'str'
    merged_df[column] = merged_df[column].astype(str)

    # Loop through characters to remove
    for character in shady_characters:
        merged_df[column] = merged_df[column].str.replace(character, "")

# Display the fixed DataFrame
merged_df
```

```python
# Drop "byline.person" column
merged_df = merged_df.drop("byline.person", axis= 1)
merged_df
```

```python
# Delete duplicate rows and reset index
merged_df = merged_df.drop_duplicates()
merged_df
```

```python
# Export data to CSV without the index
merged_df.to_csv("a_csv_of_movies.csv", index= False)
```
